```python
from skimage.io import imread
img = imread('kitten.jpg')
```

# 2. Plotting and Channels

```python
from skimage.io import imshow
import matplotlib.pyplot as plt
imshow(img)
plt.axis('off')
plt.show()
```

```python
imshow(img[:,:,0], cmap='Reds_r')
```

```python
imshow(img[:,:,2], cmap='Blues_r')
```

# 3. Point Operation

```python
from skimage.exposure import rescale_intensity
import numpy as np

img1 = rescale_intensity(img, out_range=(100,255)).astype(np.uint8)
imshow(img1)
plt.axis('off')
plt.draw()
```

```python
from skimage.exposure import adjust_gamma
img1 = adjust_gamma(img, gamma=1.5).astype(np.uint8)
imshow(img1)
plt.axis('off')
plt.draw()
```

```python
from skimage.exposure import adjust_gamma
img1 = adjust_gamma(img, gamma=0.5).astype(np.uint8)
imshow(img1)
plt.axis('off')
plt.draw()
```

# 4. Filtering

```python
from skimage.filters import gaussian
img1 = gaussian(img, sigma=3, multichannel=True)
imshow(img1)
plt.axis('off')
plt.draw()
```

```python
from skimage.filters import sobel, threshold_otsu
from skimage.color import rgb2gray
import matplotlib.pyplot as plt

img0 = rgb2gray(img)
img1 = sobel(img0)
val = threshold_otsu(img1)
img2 = img1 > val
#imshow(img2.astype('uint8'))

plt.imshow(img2, cmap='gray')
plt.axis('off')
plt.draw()
```
