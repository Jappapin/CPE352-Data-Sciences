## Network Data Lab

### Given the social network data, find the following answer

```python
import pandas as pd
import networkx as nx
import matplotlib as plt
```

```python
edgelist = pd.read_csv('dolphin_edges.csv')
graph = nx.from_pandas_edgelist(edgelist, source='From', target='To', create_using=nx.MultiDiGraph)
nx.draw(graph, with_labels=True, node_size=1000, node_color='#00AAAA')
```

```python
nx.info(graph)
```

### 1. Who are the most popular?

```python
df = pd.DataFrame(graph.in_degree, columns=['Node','Degree'])
df_sorted = df.sort_values('Degree')
df_sorted.tail(5)
```

### Answer : Web and Trigger are the most popular, because they have the highest in degree value.
### in degree value เยอะ : มีคนเข้าหาเยอะ

### 2. Pick 2 individuals, find the shortest path between them

```python
nx.shortest_path(graph,'Beak','Zap')
```

### 3. Find how many components are there in the network

### Answer : There is only one component in the network

```python
graph2 = nx.from_pandas_edgelist(edgelist, source='From', target='To')
compo = [x for x in nx.connected_components(graph2)]
len(compo)
```
